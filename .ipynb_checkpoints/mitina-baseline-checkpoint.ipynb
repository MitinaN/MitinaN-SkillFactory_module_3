```python
# Значения ID  ресторана преобразуем в число убрав 'id_'
data.Restaurant_id = data.Restaurant_id.apply(lambda x: x[3:])
data.Restaurant_id = [int(x) for x in data.Restaurant_id]

# Значения ID ресторана в TripAdvisor тоже приведем к числовому формату
data.ID_TA = data.ID_TA.apply(lambda x: x[1:])
data.ID_TA = [int(x) for x in data.ID_TA]

#Обработаем столбец Cuisine Style
data['Cuisine Style'] = data['Cuisine Style'].astype(str).apply(lambda x: None if x.strip() == '' else x)
data['Cuisine Style'] = data['Cuisine Style'].apply(lambda x: x.replace("['", '').replace("', '", ',').replace("']", ''))

# уберем лишнее в Reviews
data['Reviews'] = data['Reviews'].astype(str).apply(lambda x: None if x.strip() == '' else x)
data['Reviews'] = data['Reviews'].apply(lambda x: x.replace("[['", '').replace("', '", '|').replace('], [', '|').replace("']]", ''))
data['Reviews'] = data['Reviews'].apply(lambda x: x.replace("'", ''))

# Обработаем название ресторана из URL
data['URL_TA'] = data['URL_TA'].astype(str).apply(lambda x: None if x.strip() == '' else x)
data['URL_TA'] = data['URL_TA'].apply(lambda x: x.replace("/Restaurant_Review-", '').replace(".html", '').replace('-Reviews-', '|'))
data_m = data['URL_TA'].str.split('|', expand=True)
data['restaurant'] = data_m[1]
```

```python
def outliers(column):
    outlier1 = outlier2 = 0
    """ 
    Функция проверяет наличие выбросов через межквартильный размах
    """
    q1 = data[column].quantile(q=0.25, interpolation='midpoint')
    q3 = data[column].quantile(q=0.75, interpolation='midpoint')
    MR = round(q3 - q1, 1)
    limit1 = q1 - 3 * MR
    limit2 = q3 + 3 * MR
    if data[column].min() < limit1:
        print(f'выбросы ниже {limit1}')
    elif data[column].max() > limit2:
        print(f'выбросы выше {limit2}')
    else:
        print('выбросов нет')
    return
```

```python
def fast_mode(df, key_cols, value_col):
    """ 
    Вычисление моды колонки группируя не считая нули 

    """
    return (df.groupby(key_cols + [value_col]).size() 
              .to_frame('counts').reset_index() 
              .sort_values('counts', ascending=False) 
              .drop_duplicates(subset=key_cols)).drop(columns='counts')
```

```python
data.loc[data['Price Range'].isnull(), 'Price Range'] = data.restaurant_title.map(fast_mode(data, ['restaurant_title'], 'Price Range').set_index('restaurant_title')['Price Range'])
```

```python
# importing all necessery modules 

comment_words = '' 
stopwords = set(STOPWORDS) 

newStopWords = ['restaurant', 'food']

stopwords.update(newStopWords)

  
# iterate through the csv file 
for val in data.review_text1: 
      
    # typecaste each val to string 
    val = str(val) 
  
    # split the value 
    tokens = val.split() 
      
    # Converts each token into lowercase 
    for i in range(len(tokens)): 
        tokens[i] = tokens[i].lower() 
      
    comment_words += " ".join(tokens)+" "
  
#wordcloud = WordCloud(width = 800, height = 800, 
#                background_color ='white', 
#                stopwords = stopwords, 
#                min_font_size = 10).generate(comment_words) 
wordcloud = WordCloud(
    background_color='white',
    stopwords=stopwords,
    max_words=300,
    max_font_size=200, 
    width=1000, height=800,
    random_state=42,
).generate(" ".join(data['review_1'].astype(str)))

# plot the WordCloud image                        
plt.figure(figsize = (12, 14), facecolor = None) 
plt.imshow(wordcloud) 
plt.axis("off") 
plt.tight_layout(pad = 0) 
  
plt.show()
```

```python

from nltk.sentiment.vader import SentimentIntensityAnalyzer

SIA = SentimentIntensityAnalyzer()

# Applying Model, Variable Creation
sentiment = data.copy()
sentiment['polarity_score']=sentiment.review_1.apply(lambda x:SIA.polarity_scores(x)['compound'])
sentiment['neutral_score']=sentiment.review_1.apply(lambda x:SIA.polarity_scores(x)['neu'])
sentiment['negative_score']=sentiment.review_1.apply(lambda x:SIA.polarity_scores(x)['neg'])
sentiment['positive_score']=sentiment.review_1.apply(lambda x:SIA.polarity_scores(x)['pos'])

sentiment['sentiment']= np.nan
sentiment.loc[sentiment.polarity_score>0,'sentiment']='POSITIVE'
sentiment.loc[sentiment.polarity_score==0,'sentiment']='NEUTRAL'
sentiment.loc[sentiment.polarity_score<0,'sentiment']='NEGATIVE'
data['sentiment'] = sentiment.sentiment
```

# Data Preprocessing
Теперь, для удобства и воспроизводимости кода, завернем всю обработку в одну большую функцию.

```python
# на всякий случай, заново подгружаем данные
df_train = pd.read_csv(DATA_DIR+'/main_task.csv')
df_test = pd.read_csv(DATA_DIR+'/kaggle_task.csv')
df_train['sample'] = 1 # помечаем где у нас трейн
df_test['sample'] = 0 # помечаем где у нас тест
df_test['Rating'] = 0 # в тесте у нас нет значения Rating, мы его должны предсказать, по этому пока просто заполняем нулями

data = df_test.append(df_train, sort=False).reset_index(drop=True) # объединяем
data.info()
```

```python
def fast_mode(df, key_cols, value_col):
    """ 
    Вычисление моды колонки 

    """
    return (df.groupby(key_cols + [value_col]).size() 
              .to_frame('counts').reset_index() 
              .sort_values('counts', ascending=False) 
              .drop_duplicates(subset=key_cols)).drop(columns='counts')
```

```python
def preproc_data(df_input):
    '''includes several functions to pre-process the predictor data.'''
    
    df_output = df_input.copy()
    
    # ################### 1. Предобработка ############################################################## 
     
    df_output['Reviews'] = df_output['Reviews'].astype(str).apply(lambda x: None if x.strip() == '' else x)
    df_output['Reviews'] = df_output['Reviews'].apply(lambda x: x.replace("[['", '').replace("', '", '|').replace('], [', '|').replace("']]", '')) #(lambda x: x.replace("['", '') if ']' in x else x))
    df_output['Reviews'] = df_output['Reviews'].apply(lambda x: x.replace("'", ''))
    df_output.ID_TA = df_output.ID_TA.apply(lambda x: x[1:])
    df_output.ID_TA = [int(x) for x in df_output.ID_TA]

    df_output['URL_TA'] = df_output['URL_TA'].astype(str).apply(lambda x: None if x.strip() == '' else x)
    df_output['URL_TA'] = df_output['URL_TA'].apply(lambda x: x.replace("/Restaurant_Review-", '').replace(".html", '').replace('-Reviews-', '|'))
    data_m = df_output['URL_TA'].str.split('|', expand=True)
    df_output['restaurant'] = data_m[1]
    data_a = df_output['restaurant'].str.split('-', expand=True)
    df_output['restaurant_title'] = data_a[0]
    df_output['country'] = data_a[1]
    df_output.country = df_output.country.fillna('Athens')
    df_output = df_output.assign(rest_length3 = df_output['country'].apply(len))
    
    df_output = df_output.assign(rest_length = df_output['restaurant'].apply(len))
    df_output = df_output.assign(rest_length2 = df_output['restaurant_title'].apply(len))


    df_output.loc[df_output['Price Range'].isnull(), 'Price Range'] = df_output.restaurant_title.map(fast_mode(df_output, ['restaurant_title'], 'Price Range').set_index('restaurant_title')['Price Range'])
    df_output['Price Range'] = df_output['Price Range'].astype(str).apply(lambda x: x.replace("$$ - $$$", '2').replace('$$$$', '3').replace("$", '1'))
    # ################### 2. NAN ############################################################## 
    # Далее заполняем пропуски, вы можете попробовать заполнением средним или средним по городу и тд...
    df_output['Number of Reviews'].fillna(0, inplace=True)
    df_output['Price Range'].fillna(2, inplace=True)
    
    #Обработаем столбец Cuisine Style
    df_output['Cuisine Style'] = df_output['Cuisine Style'].astype(str).apply(lambda x: None if x.strip() == '' else x)
    df_output['Cuisine Style'] = df_output['Cuisine Style'].apply(lambda x: x.replace("['", '').replace("', '", ',').replace("']", ''))

    df_output['Cuisine Style'] = df_output['Cuisine Style'].apply(lambda x: x.replace('nan', 'other'))
    df_output['cuisine_len'] = df_output['Cuisine Style'].apply(lambda x: len(x.split(',')))
    df_output['bar'] = df_output['Cuisine Style'].apply(lambda x: 1 if ('Bar' or  'Pub') in x else 0)
    df_output['fast_food'] = df_output['Cuisine Style'].apply(lambda x: 1 if ('Cafe' or  'Fast food') in x else 0)
    df_output['Vegetarian Friendly'] = df_output['Cuisine Style'].apply(lambda x: 1 if 'Vegetarian Friendly' in x else 0)
    df_output['Vegan Options'] = df_output['Cuisine Style'].apply(lambda x: 1 if 'Vegan Options' in x else 0)
    df_output['Gluten Free Options'] = df_output['Cuisine Style'].apply(lambda x: 1 if 'Gluten Free Options' in x else 0)
    df_output['Italian'] = df_output['Cuisine Style'].apply(lambda x: 1 if 'Italian' in x else 0)
    
                                                                                      
    df_output['Pizza'] = df_output['Cuisine Style'].apply(lambda x: 1 if 'Pizza' in x else 0)
    df_output['European'] = df_output['Cuisine Style'].apply(lambda x: 1 if ('European') in x else 0)
    df_output['Mediterranean'] = df_output['Cuisine Style'].apply(lambda x: 1 if ('Mediterranean') in x else 0)
    df_output['Italian'] = df_output['Cuisine Style'].apply(lambda x: 1 if ('Italian' or  'Pizza') in x else 0)
    df_output['French'] = df_output['Cuisine Style'].apply(lambda x: 1 if ('French') in x else 0)
                                                                                      
    
    df_output['Asian'] = df_output['Cuisine Style'].apply(lambda x: 1 if ('Asian') in x else 0)
    df_output['seafood'] = df_output['Cuisine Style'].apply(lambda x: 1 if ('Sushi' or  'Japanese' or 'Seafood') in x else 0)

    data_reviews = df_output['Reviews'].str.split('|', expand=True)
    data_reviews.columns = ['review_1', 'review_2', 'review_date_1', 'review_date_2', 'nen']
    data_reviews
    data_reviews['review_date_1'] = data_reviews['review_date_1'].astype(str).apply(lambda x: x.replace('Pub/Restaurant', '07/19/2017'))
    df_output['review_1'] = data_reviews.review_1
    df_output['review_2'] = data_reviews['review_2'].astype(str).apply(lambda x: x.replace('09/24/2017', ''))
    #df_output['review_2'] = data_reviews.review_2
    data_reviews['review_date_1'] = data_reviews['review_date_1'].astype(str).apply(lambda x: None if x.strip() == '' else x) 
    df_output['review_date_1'] = data_reviews['review_date_1'].apply(lambda x: x.replace('None', '01/07/2018'))
    data_reviews['review_date_2'] = data_reviews['review_date_2'].astype(str).apply(lambda x: None if x.strip() == '' else x) 
    df_output['review_date_2'] = data_reviews['review_date_2'].apply(lambda x: x.replace('None', '01/03/2018'))
    
    df_output['review_date_1']= pd.to_datetime(df_output['review_date_1'])
    df_output['review_date_2']= pd.to_datetime(df_output['review_date_2'])
    df_output['year_1'] = df_output['review_date_1'].dt.year
    df_output['year_2'] = df_output['review_date_2'].dt.year
    df_output['days'] = (df_output.review_date_2 - df_output.review_date_1)
    df_output['days'] = df_output['days'].apply(lambda x: abs(x))

    df_output['days'] = df_output['days'].dt.days

   
    # Добавим текущую дату
    df_output['current_time'] = pd.to_datetime('today').strftime("%m/%d/%Y")
    # Заполним пропуски
    df_output.year_1 = df_output.year_1.fillna(df_output.year_1.mean)
    df_output.year_2 = df_output.year_2.fillna(df_output.year_2.mean)
    #
    df_output['current_time'] = pd.to_datetime(df_output['current_time'])
    df_output['review_date_1']= pd.to_datetime(df_output['review_date_1'])

    df_output['review_date_2']= pd.to_datetime(df_output['review_date_2'])
    #
    df_output['timedelta_1'] = (df_output['current_time'] - df_output['review_date_1']).astype('timedelta64[Y]')
    df_output['timedelta_2'] = (df_output['current_time'] - df_output['review_date_2']).astype('timedelta64[Y]')

    df_output['review_text1'] = df_output['Reviews'].apply(lambda x: x.lower())
    df_output['no_review'] = df_output.review_text1.apply(lambda x: 0 if x == '[[|]]' else 1)

    df_output['not bad'] = df_output.review_text1.apply(lambda x: 1 if 'not bad' in x else 0)
    df_output.review_text1 = df_output.review_text1.apply(lambda x: x.replace('not bad', 'good'))


    df_output['good'] = df_output.review_text1.apply(lambda x: 1 if 'good' in x else 0)
    df_output['great'] = df_output.review_text1.apply(lambda x: 1 if 'great' in x else 0)
    df_output['tasty'] = df_output.review_text1.apply(lambda x: 1 if 'tasty' in x else 0)
    df_output['excellent'] = df_output.review_text1.apply(lambda x: 1 if 'excellent' in x else 0)
    df_output['terrible'] = df_output.review_text1.apply(lambda x: 1 if 'terrible' in x else 0)
    df_output['service'] = df_output.review_text1.apply(lambda x: 1 if 'service' in x else 0)
    df_output['bad'] = df_output.review_text1.apply(lambda x: 1 if 'bad' in x else 0)
    df_output['excellent'] = df_output.review_text1.apply(lambda x: 1 if 'excellent' in x else 0)
    
    df_output['experiense'] = df_output.review_text1.apply(lambda x: 1 if 'experiense' in x else 0)
    df_output['amazing'] = df_output.review_text1.apply(lambda x: 1 if 'amazing' in x else 0)
    df_output['fantastic'] = df_output.review_text1.apply(lambda x: 1 if 'fantastic' in x else 0)
    df_output['deliciou'] = df_output.review_text1.apply(lambda x: 1 if 'deliciou' in x else 0)
    df_output['average'] = df_output.review_text1.apply(lambda x: 1 if 'average' in x else 0)
    df_output['best'] = df_output.review_text1.apply(lambda x: 1 if 'best' in x else 0)
    
    
         
    # ################### 3. Encoding ############################################################## 
    # для One-Hot Encoding в pandas есть готовая функция - get_dummies. Особенно радует параметр dummy_na
    
  
    
    # для One-Hot Encoding в pandas есть готовая функция - get_dummies. Особенно радует параметр dummy_na
    means = df_output.groupby('City').Ranking.mean()
    std = df_output.groupby('City').Ranking.std()
    
    df_output['Ranking_std'] = (df_output.Ranking - df_output.City.map(means)) / df_output.City.map(std)
 
    means1 = df_output.groupby('City')['Number of Reviews'].mean()
    std1 = df_output.groupby('City')['Number of Reviews'].std()
    df_output['number_reviews_std'] = (df_output['Number of Reviews'] - df_output['City'].map(means1)) / df_output['City'].map(std1)
   
    means1 = df_output.groupby('Price Range')['Ranking'].mean()
    std1 = df_output.groupby('Price Range')['Ranking'].std()
    df_output['number_reviews-PR_std'] = (df_output['Ranking'] - df_output['Price Range'].map(means1)) / df_output['Price Range'].map(std1)
   
    # ################### 4. Feature Engineering ####################################################
    
    df_output = pd.get_dummies(df_output, columns=[ 'City'], dummy_na=True)
    
    # ################### 5. Clean #################################################### 
    # убираем признаки которые еще не успели обработать, 
    # модель на признаках с dtypes "object" обучаться не будет, просто выберим их и удалим
    object_columns = [s for s in df_output.columns if df_output[s].dtypes == 'object']
    df_output.drop(object_columns, axis = 1, inplace=True)
    df_output.drop(['current_time','review_date_1', 'review_date_2',], axis = 1, inplace=True)
    return df_output
   
```

```python
# Теперь выделим тестовую часть
train_data = df_preproc.query('sample == 1').drop(['sample'], axis=1)
test_data = df_preproc.query('sample == 0').drop(['sample'], axis=1)

y = train_data.Rating.values            # наш таргет
X = train_data.drop(['Rating'], axis=1)
```

**Перед тем как отправлять наши данные на обучение, разделим данные на еще один тест и трейн, для валидации.

```python
# Воспользуемся специальной функцие train_test_split для разбивки тестовых данных

# выделим 20% данных на валидацию (параметр test_size)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=RANDOM_SEED)
```

```python
# проверяем
test_data.shape, train_data.shape, X.shape, X_train.shape, X_test.shape
```

# Model 
Сам ML

```python
# Импортируем необходимые библиотеки:
from sklearn.ensemble import RandomForestRegressor # инструмент для создания и обучения модели
from sklearn import metrics # инструменты для оценки точности модели
```

```python
# Создаём модель (НАСТРОЙКИ НЕ ТРОГАЕМ)
model = RandomForestRegressor(n_estimators=100, verbose=1, n_jobs=-1, random_state=RANDOM_SEED)
```

```python
# Обучаем модель на тестовом наборе данных
model.fit(X_train, y_train)

# Используем обученную модель для предсказания рейтинга ресторанов в тестовой выборке.
# Предсказанные значения записываем в переменную y_pred
y_pred = model.predict(X_test)
```

```python
y = pd.DataFrame(y_pred)
```

```python
#y = pd.DataFrame(y_pred)
y['okr'] = round(y[0])
#y [0] = y[0].apply(lambda x: x if )
y['r'] = y['okr'] - y[0]
#y = y.apply(lambda x: round (x, 1))
y['r'] = y.r.apply(lambda x: 0 if abs(x) <= 0.25 else (0.5 if x < 0 else -0.5))
y['new'] = y['okr'] + y['r']
y.head(10)
```

```python
y_pred = np.array(y['new'])
y_pred
```

```python
# Сравниваем предсказанные значения (y_pred) с реальными (y_test), и смотрим насколько они в среднем отличаются
# Метрика называется Mean Absolute Error (MAE) и показывает среднее отклонение предсказанных значений от фактических.
print('MAE:', metrics.mean_absolute_error(y_test, y_pred))
```

###### MAE: 0.16325

```python
# в RandomForestRegressor есть возможность вывести самые важные признаки для модели
plt.rcParams['figure.figsize'] = (10,10)
feat_importances = pd.Series(model.feature_importances_, index=X.columns)
feat_importances.nlargest(70).plot(kind='barh')
```
